```python
df = pd.read_csv(r"C:\Users\tuka3\OneDrive\Escritorio\UPGRADE\Temario\M6.Graficas\Plotly\ds_salaries.csv")
```

```python
fig = px.histogram(df, x="salary", nbins=50, title="Histograma de Salarios")
fig.show()
```

```python
fig= px.box(df, x="salary", title="Boxplot de Salarios")
fig.show()
```

```python
fig= px.density_heatmap(df, x="job_title", y="employment_type", marginal_x="histogram", marginal_y="histogram", template="simple_white")
fig.show()
```

```python
fig = px.scatter_3d(df, x="salary", y="job_title", z="employment_type", color="job_title", size="salary", hover_name="job_title", symbol="employment_type", title="Salarios por Puesto de Trabajo y Tipo de Empleo")
fig.show()
```

```python
fig = px.scatter(df, x="salary", y="job_title", color="employment_type", size="salary", hover_name="job_title", title="Salarios por Puesto de Trabajo y Tipo de Empleo")
fig.show()
```

```python
fig = px.treemap(df, path=["job_title", "employment_type"], values="salary", title="Treemap de Salarios por Puesto de Trabajo y Tipo de Empleo")
fig.show()
```
